```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("Modules imported successfully")
```

```python
corono_dataset_csv = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
corono_dataset_csv.head(10)
```

```python
corono_dataset_csv.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
corono_dataset_csv_aggregated=corono_dataset_csv.groupby("Country/Region").sum()
```

```python
country = list(corono_dataset_csv_aggregated.index)
max_infection_rate =[]
for c in country:
    max_infection_rate.append(corono_dataset_csv_aggregated.loc[c].diff().max())
```

```python
corono_dataset_csv_aggregated["max_infection_rate"] = max_infection_rate
```

```python
corono_dataset_csv_aggregated.head(10)
```

```python
corono_data = pd.DataFrame(corono_dataset_csv_aggregated["max_infection_rate"])
```
